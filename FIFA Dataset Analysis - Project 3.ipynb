## FIFA Dataset Analysis

```python
#Loading the libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv('players_20.csv')
```

```python
fifa.head()
```

```python
# Printing all columns using for loop
for col in fifa.columns:
  print(col)
```

```python
# No . of rows and columns
fifa.shape
```

```python
# Countries specifications
fifa['nationality'].value_counts()
```

```python
# Top 10 countries where most of the players play fifa
fifa['nationality'].value_counts()[0:10]
```

```python
# Top five
fifa['nationality'].value_counts()[0:5]
```

```python
# Executing names without values
fifa['nationality'].value_counts()[0:5].keys()
```

```python
# Plotting a bar plot for top 5 countries where most of the players play fifa
plt.figure(figsize = (8,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color = ['red','green','blue','yellow','violet'])
plt.show()
```

```python
# Finding players salaries
player_salary = fifa[['short_name','wage_eur']]
```

```python
player_salary.head()
```

```python
# Finding top players salaries in descending order
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
player_salary.head()
```

```python
# Plotting a bar plot for players with highest salaries
plt.figure(figsize = (8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color= ['yellow','blue','pink','orange','lime' ])
plt.show()
```

```python
# Analyzing players who belong to germany
fifa['nationality'] == 'Germany'
```

```python
# Germany
Germany = fifa[fifa['nationality'] == 'Germany']
Germany.head(10)
```

```python
# Viewing tallest german player
Germany.sort_values(by = ['height_cm'],ascending = False).head(10)
```

```python
# Viewimg player with more weight from germany
Germany.sort_values(by = ['weight_kg'],ascending = False).head(10)
```

```python
# Highest paid german player
Germany.sort_values(by = ['wage_eur'],ascending = False).head(10)
```

```python
# Names by wages of german players
Germany[['short_name','wage_eur']].sort_values(by = ['wage_eur'],ascending = False).head(10)
```
